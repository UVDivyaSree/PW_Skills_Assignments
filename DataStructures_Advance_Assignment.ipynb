#2.Write a code to count the number of vowels in a string.

```python
def count_vowels(input_string):
    # Define a set of vowels
    vowels = set("aeiouAEIOU")
    # Initialize a counter
    count = 0

    # Iterate through each character in the input string
    for char in input_string:
        # If the character is a vowel, increment the counter
        if char in vowels:
            count += 1

    return count

# Example usage
string = "Hello, World!"
vowel_count = count_vowels(string)
print(f"The number of vowels in '{string}' is: {vowel_count}")
```

#3.Write a code to check if a given string is a palindrome or not.

```python
def is_palindrome(input_string):
    # Convert the string to lowercase and remove non-alphanumeric characters
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())
    # Check if the cleaned string is equal to its reverse
    return cleaned_string == cleaned_string[::-1]

# Example usage
string = "Noon"
if is_palindrome(string):
    print(f"'{string}' is a palindrome.")
else:
    print(f"'{string}' is not a palindrome.")
```

#4.Write a code to check if two given strings are anagrams of each other.



```python
def are_anagrams(string1, string2):
    # Remove spaces, convert to lowercase, and sort the characters of both strings
    cleaned_string1 = sorted(string1.replace(" ", "").lower())
    cleaned_string2 = sorted(string2.replace(" ", "").lower())

    # Check if the sorted versions of both strings are equal
    return cleaned_string1 == cleaned_string2

# Example usage
str1 = "Listen"
str2 = "Silent"

if are_anagrams(str1, str2):
    print(f"'{str1}' and '{str2}' are anagrams.")
else:
    print(f"'{str1}' and '{str2}' are not anagrams.")
```

#5.Write a code to find all occurrences of a given substring within another string.

```python
def find_all_occurrences(main_string, substring):
    # Initialize an empty list to store the positions of occurrences
    occurrences = []
    # Initialize the start index
    start = 0

    # Loop to find all occurrences of the substring
    while True:
        # Find the index of the substring in the main string starting from 'start'
        index = main_string.find(substring, start)

        # If the substring is found, add the index to the list
        if index != -1:
            occurrences.append(index)
            # Update the start index to search for the next occurrence
            start = index + 1
        else:
            # Break the loop if no more occurrences are found
            break

    return occurrences

# Example usage
main_string = "This is a test string, and this is another test string."
substring = "test"

occurrences = find_all_occurrences(main_string, substring)
print(f"Occurrences of '{substring}' in the main string: {occurrences}")
```

#6.Write a code to perform basic string compression uisng the counts of repeated characters.

```python
def compress_string(input_string):
    # Initialize variables to store the compressed string and the count of characters
    compressed = []
    count = 1

    # Iterate through the string characters
    for i in range(1, len(input_string)):
        # If the current character is the same as the previous one, increment the count
        if input_string[i] == input_string[i - 1]:
            count += 1
        else:
            # Otherwise, add the previous character and its count to the compressed list
            compressed.append(input_string[i - 1] + str(count))
            # Reset the count for the new character
            count = 1

    # Add the last character and its count
    compressed.append(input_string[-1] + str(count))

    # Join the compressed list into a single string
    compressed_string = ''.join(compressed)

    # Return the original string if compression does not reduce the length
    return compressed_string if len(compressed_string) < len(input_string) else input_string

# Example usage
input_string = "aabcccccaaa"
compressed = compress_string(input_string)
print(f"Compressed string: {compressed}")
```

#7.Write a code to determine if a string has all unique characters.

```python
def has_all_unique_characters(input_string):
    # Create an empty set to store characters
    char_set = set()

    # Iterate through each character in the input string
    for char in input_string:
        # If the character is already in the set, it means it's not unique
        if char in char_set:
            return False
        # Add the character to the set
        char_set.add(char)

    # If no duplicate character is found, return True
    return True

# Example usage
input_string = "abcdefg"
result = has_all_unique_characters(input_string)
print(f"Does '{input_string}' have all unique characters? {result}")
```

#12.Implement a code to find the second largest number in  a given list of intergers.

```python
def find_second_largest(nums):
    # Remove duplicates by converting the list to a set
    nums = list(set(nums))

    if len(nums) < 2:
        return "List must contain at least two unique numbers."

    # Sort the list in descending order
    nums.sort(reverse=True)

    # Return the second element (second largest number)
    return nums[1]

# Example usage
numbers = [10, 20, 4, 45, 99, 99, 20]
second_largest = find_second_largest(numbers)
print(f"The second largest number is: {second_largest}")
```

#14.Write a code to reverse a list in-place without using any built-in reverse functions.

```python
def reverse_list(lst):
    # Initialize two pointers, one at the beginning and one at the end
    start = 0
    end = len(lst) - 1

    # Swap elements while start is less than end
    while start < end:
        # Swap the elements
        lst[start], lst[end] = lst[end], lst[start]

        # Move the pointers towards each other
        start += 1
        end -= 1

# Example usage
elements = [10, 20, 30, 40, 50]
print(f"Original list: {elements}")
reverse_list(elements)
print(f"Reversed list: {elements}")
```

#15.Implement a code to find and remove duplicates from a list while preserving the original order of elements.

```python
def remove_duplicates(lst):
    # Initialize an empty list to store unique elements
    unique_list = []
    # Initialize a set to track seen elements
    seen = set()

    # Loop through each element in the list
    for item in lst:
        # If the item has not been seen before, add it to the unique list and mark it as seen
        if item not in seen:
            unique_list.append(item)
            seen.add(item)

    return unique_list

# Example usage
elements = [10, 20, 10, 30, 40, 50, 20, 30]
unique_elements = remove_duplicates(elements)
print(f"Original list: {elements}")
print(f"List after removing duplicates: {unique_elements}")
```

#16.Create a code to check if a given list is sorted(either in ascending or descending order) or not.

```python
def is_sorted(lst):
    # Check if the list is sorted in ascending order
    ascending = all(lst[i] <= lst[i + 1] for i in range(len(lst) - 1))

    # Check if the list is sorted in descending order
    descending = all(lst[i] >= lst[i + 1] for i in range(len(lst) - 1))

    # Return the result
    if ascending:
        return "List is sorted in ascending order."
    elif descending:
        return "List is sorted in descending order."
    else:
        return "List is not sorted."

# Example usage
elements1 = [10, 20, 30, 40, 50]
elements2 = [50, 40, 30, 20, 10]
elements3 = [10, 50, 30, 20, 40]

print(is_sorted(elements1))  # Ascending
print(is_sorted(elements2))  # Descending
print(is_sorted(elements3))  # Not sorted
```

#17.Write a code to merge two sorted lists into a single sorted list.

```python
def merge_sorted_lists(list1, list2):
    # Initialize pointers for both lists
    i, j = 0, 0
    # Initialize an empty list to store the merged result
    merged_list = []

    # Traverse both lists and merge them
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # If any elements are left in list1, add them
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # If any elements are left in list2, add them
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]

merged_list = merge_sorted_lists(sorted_list1, sorted_list2)
print(f"Merged sorted list: {merged_list}")
```

#20.Write  a code to shuffle a given list randomly without using any built-in functions.

```python
import random

def shuffle_list(lst):
    # Get the length of the list
    n = len(lst)

    # Iterate over the list
    for i in range(n):
        # Generate a random index between the current index and the last index
        j = random.randint(i, n - 1)

        # Swap the current element with the randomly chosen element
        lst[i], lst[j] = lst[j], lst[i]

    return lst

# Example usage
elements = [1, 2, 3, 4, 5]
print(f"Original list: {elements}")
shuffled_list = shuffle_list(elements)
print(f"Shuffled list: {shuffled_list}")
```

#31.Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list.

```python
def word_frequencies(words):
    frequency_dict = {}

    for word in words:
        # Convert the word to lower case to ensure case-insensitivity
        word = word.lower()
        if word in frequency_dict:
            # Increment the count if the word is already in the dictionary
            frequency_dict[word] += 1
        else:
            # Add the word to the dictionary with count 1 if it is not present
            frequency_dict[word] = 1

    return frequency_dict

# Example usage
words_list = ['apple', 'banana', 'Apple', 'orange', 'banana', 'apple', 'Orange']
frequency_dict = word_frequencies(words_list)
print(f"Word frequencies: {frequency_dict}")
```

#32.Write a code that takes two dictionaries as input and merges them into a single dictionary.If there are common keys,the values should be added together.

```python
def merge_dictionaries(dict1, dict2):
    # Create a new dictionary to store the merged result
    merged_dict = {}

    # Add all key-value pairs from the first dictionary
    for key, value in dict1.items():
        merged_dict[key] = value

    # Add all key-value pairs from the second dictionary
    for key, value in dict2.items():
        if key in merged_dict:
            # If the key already exists, add the values
            merged_dict[key] += value
        else:
            # Otherwise, add the new key-value pair
            merged_dict[key] = value

    return merged_dict

# Example usage
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 5, 'c': 10, 'd': 15}

merged_dict = merge_dictionaries(dict1, dict2)
print(f"Merged dictionary: {merged_dict}")
```

#33.Write a code to access a value in a nested dictionary.The function should take the dictionary and a list of keys as input and return the corresponding value.If any keys do not exist in the dictionary, the function should return None.

```python
def access_nested_value(d, keys):
    # Initialize current dictionary level
    current = d

    for key in keys:
        # Check if the key exists in the current dictionary level
        if isinstance(current, dict) and key in current:
            current = current[key]
        else:
            # Return None if key does not exist
            return None

    return current

# Example usage
nested_dict = {
    'a': {
        'b': {
            'c': 42
        }
    },
    'x': {
        'y': 100
    }
}

# Access value using a list of keys
keys = ['a', 'b', 'c']
value = access_nested_value(nested_dict, keys)
print(f"Value at keys {keys}: {value}")

# Access value with non-existent key
keys = ['a', 'b', 'z']
value = access_nested_value(nested_dict, keys)
print(f"Value at keys {keys}: {value}")

# Access value using another set of keys
keys = ['x', 'y']
value = access_nested_value(nested_dict, keys)
print(f"Value at keys {keys}: {value}")
```

#34.Write a code that takes a dictionary as input and returns a sorted version of it based on the values.You can choose whether to sort in ascending or descending order.

```python
def sort_dictionary_by_values(d, reverse=False):
    # Sort the dictionary by values and return a new dictionary
    sorted_dict = dict(sorted(d.items(), key=lambda item: item[1], reverse=reverse))

    return sorted_dict

# Example usage
input_dict = {
    'apple': 3,
    'banana': 1,
    'cherry': 2,
    'date': 4
}

# Sort in ascending order
sorted_dict_asc = sort_dictionary_by_values(input_dict)
print(f"Dictionary sorted by values (ascending): {sorted_dict_asc}")

# Sort in descending order
sorted_dict_desc = sort_dictionary_by_values(input_dict, reverse=True)
print(f"Dictionary sorted by values (descending): {sorted_dict_desc}")
```

#35.Write a code that inverts a dictionary,swapping keys and values.Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

```python
def invert_dictionary(original_dict):
    inverted_dict = {}

    for key, value in original_dict.items():
        # If the value is already a key in the inverted dictionary, append the key to the list of keys
        if value in inverted_dict:
            if isinstance(inverted_dict[value], list):
                inverted_dict[value].append(key)
            else:
                inverted_dict[value] = [inverted_dict[value], key]
        else:
            # If the value is not in the inverted dictionary, create a new entry with the current key
            inverted_dict[value] = key

    return inverted_dict

# Example usage
original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3
}

inverted_dict = invert_dictionary(original_dict)
print(f"Original dictionary: {original_dict}")
print(f"Inverted dictionary: {inverted_dict}")
```
