#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
size = 100
feature1 = np.random.normal(50, 10, size)
feature2 = feature1 * 0.8 + np.random.normal(0, 5, size)  # Strongly correlated with feature1
feature3 = feature1 * -0.5 + feature2 * 0.3 + np.random.normal(0, 7, size)  # Moderately correlated
feature4 = np.random.normal(30, 8, size)  # Weakly correlated

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': feature1,
    'Feature 2': feature2,
    'Feature 3': feature3,
    'Feature 4': feature4
})

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)

# Add labels and title
plt.title('Correlation Matrix Heatmap', fontsize=14)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z using the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data to create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Create the surface plot
ax.plot_surface(x, y, z, cmap='viridis')

# Set plot labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$Z = \sin(\sqrt{x^2 + y^2})$', fontsize=14)

# Show the plot
plt.show()
````

#3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource

# Enable output in Jupyter Notebook
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data={'fruits': fruits, 'counts': counts})

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Counts", width=800, height=400)

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.8, source=source, color="skyblue", legend_label="Fruit Counts")

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p, notebook_handle=True)
````

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
import numpy as np

# Enable output in Jupyter Notebook
output_notebook()

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a histogram plot
p = figure(title="Histogram of Data", x_axis_label="Value", y_axis_label="Frequency", width=800, height=400)

# Add quads to represent the histogram
p.quad(
    top=hist,
    bottom=0,
    left=edges[:-1],
    right=edges[1:],
    fill_color="skyblue",
    line_color="white",
    alpha=0.7,
    legend_label="Frequency"
)

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Show the plot
show(p, notebook_handle=True)
````

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.io import push_notebook
import numpy as np

# Enable output in Jupyter Notebook
output_notebook()

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Flatten the grid for Bokeh data source
xx, yy = np.meshgrid(x, y)
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

# Create a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=values.min(), high=values.max())

# Create a heatmap plot
p = figure(
    title="Heatmap Example",
    x_axis_label="X-axis",
    y_axis_label="Y-axis",
    tools="hover",
    tooltips=[("Value", "@image")],
    width=600,
    height=600
)

# Add image glyph
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title="Value")
p.add_layout(color_bar, 'right')

# Show the heatmap
show(p, notebook_handle=True)
````
